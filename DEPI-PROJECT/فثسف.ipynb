```python
import chromadb

def test_chroma():
    try:
        print("üîç Testing ChromaDB (v0.4.x style)...")

        client = chromadb.PersistentClient(path="./chroma_test")

        collection = client.get_or_create_collection(
            name="test_collection",
            metadata={"hnsw:space": "cosine"}
        )
        print("‚úÖ Collection created")

        collection.add(
            ids=["1", "2"],
            documents=["hello world", "machine learning"],
            embeddings=[[0.1, 0.2, 0.3], [0.9, 0.8, 0.7]]
        )
        print("‚úÖ Embeddings inserted")

        result = collection.query(
            query_embeddings=[[0.1, 0.2, 0.25]],
            n_results=2
        )
        print("‚úÖ Query executed")
        print("üîé Result:", result)

        print("\nüéâ ChromaDB WORKS SUCCESSFULLY!\n")

    except Exception as e:
        print("\n‚ùå ERROR OCCURRED\n")
        print(e)

if __name__ == "__main__":
    test_chroma()
```
